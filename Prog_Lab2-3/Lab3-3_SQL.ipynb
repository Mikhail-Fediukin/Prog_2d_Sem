```python
#2.1 
cur.execute(
    '''select distinct music_genre, 
round(count(music_genre) * 100 / (select count(*) from music_genre), 2)
from music_genre where not music_genre = ''
group by music_genre'''
)
#cur.fetchall()
```

```python
#2.3
cur.execute(
    '''select music_genre, avg(danceability) as dance_ability
from music_genre where not music_genre = ''
group by music_genre
order by dance_ability desc;''' 
)
#cur.fetchall()
```

```python
#2.4 
cur.execute(
    '''select music_genre, "mode", count("mode") 
from music_genre where not music_genre = ''
group by music_genre, "mode"
order by music_genre;''' 
)
#cur.fetchall()
```

```python
#8.1
cur.execute(
'''update music_genre
set song_length =
        case when duration_ms <= 180000
            then "короткая"
            when duration_ms > 300000
            then "длинная"
            else "средняя"
        end;'''
)
cur.fetchall()
```
